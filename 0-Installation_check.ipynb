## Data loading checks
Okay, let's check if the data can be found. The default path is within the folder where the notebooks are, that's the assumption of the following notebooks. You can of course change also this.

```python
from pathlib import Path
import mne

data_path = Path("./ds000117_pruned")  # current directory / ds000117_pruned
extra_path = Path("./extra_data_mne")  # current directory / extra_data_mne

print(f"checking if data path exists: {data_path.resolve()}")
print(data_path.exists())
print(f"sub-01 directory exists: {(data_path / 'sub-01').exists()}")

print(f"checking if extra path exists: {extra_path.resolve()}")
print(extra_path.exists())
```

Based on the `data_path` you specified above, this is where the raw MEG data should be, let's check if the data can be loaded using MNE:

```python
meg_folder = data_path / "derivatives" / "meg_derivatives" / "sub-01" / "ses-meg" / "meg" 
raw_fname = meg_folder / "sub-01_ses-meg_task-facerecognition_run-01_proc-sss_meg.fif"
print(raw_fname)

raw = mne.io.read_raw(raw_fname, preload=False)
print(raw)
if type(raw) is mne.io.fiff.raw.Raw:
    print("data successfully loaded!")
```

Let's check if display of 3D plotting & GUI is reasonable, this should open a window with a deformed head and some colored dots. It's not necessary to grasp what this all means yet, just verify that this window pops up.

```python
subjects_dir = extra_path / "freesurfer"
mne.gui.coregistration(subject='sub-01', subjects_dir=subjects_dir, inst=raw_fname);
```

```python
from shutil import copyfile
for surface_name in ["brain", "inner_skull", "outer_skull", "outer_skin"]:
    copyfile(f"{extra_path}/freesurfer/sub-01/bem/watershed/sub-01_{surface_name}_surface",
             f"{extra_path}/freesurfer/sub-01/bem/{surface_name}.surf")
```
